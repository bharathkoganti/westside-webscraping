```python
import requests
```

```python
import bs4
```

```python
url="https://www.westside.com/collections/men-new-in"
```

```python
data=requests.get(url)
data
```

```python
type(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(data.text,"html")
```

```python
soup.find('div',attrs={"class":"boost-pfs-filter-product-bottom-inner"}).text
```

```python
soup.find('span',attrs={'class':"boost-pfs-filter-product-item-regular-price"}).text
```

```python
a=soup.find_all('div',attrs={"class":"boost-pfs-filter-product-bottom-inner"})
Brand=[x.text for x in a]
print(Brand)
```

```python
b=soup.find_all('span',attrs={'class':"boost-pfs-filter-product-item-regular-price"})
Price=[x.text for x in b]
print(Price)
```

```python
import pandas as pd
```

```python
h=pd.DataFrame({"Brand":Brand,"Price":Price})
h
```

```python
h.to_csv("WestSidee.csv")
```
